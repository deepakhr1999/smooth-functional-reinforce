```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from main import CONFIGS
sns.set()
```

```python
# cfg = "medium"
for cfg in ["tiny", "small"]:
    grid_size = CONFIGS[cfg]["size"]
    rewards = np.zeros((grid_size, grid_size))
    goal = [grid_size - 1, grid_size - 1]
    for idx in range(grid_size):
        for jdx in range(grid_size):
            rewards[idx, jdx] = -(sum(goal) - idx - jdx) / (5 * grid_size)
    heat = np.log1p(rewards - rewards.min())
    heat[-1, -1] = 0.75
    rewards[-1, -1] = grid_size ** 2
    fig, ax = plt.subplots(figsize=(8, 8))
    sns.heatmap(heat, annot=rewards, fmt=".2f", cbar=False, ax=ax, cmap="rocket")
    plt.title(f"Rewards for {cfg} grid", fontsize=14)
    plt.savefig(f"images/grid_{cfg}.png", bbox_inches="tight")
    plt.show()

```
